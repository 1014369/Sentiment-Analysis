```python
import re
import string
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import pandas as pd
import numpy as np
```

```python
nltk.download("stopwords")
nltk.download("wordnet")
nltk.download("punkt")
```

```python
df=pd.read_csv("C:/Users/user/Desktop/Sentiment analysis of Financial News/Data/EDA",encoding="ISO-8859-1")
pd.set_option("display.max_colwidth",None)
df.head()
```

```python
df["Sentence"]=df["Sentence"].str.lower()
```

```python
def clean_text(text):
    text=re.sub(r'https\S|www\S+',"",text)
    text=re.sub(r'\d+',"",text)
    text=text.translate(str.maketrans("","",string.punctuation))
    return text
```

```python
df["Sentence"] = df["Sentence"].apply(clean_text)
```

```python
df["tokens"] = df["Sentence"].apply(nltk.word_tokenize)
```

```python
stop_words=set(stopwords.words("english"))
df["tokens"]=df["tokens"].apply(
    lambda x: [word for word in x if word not in stop_words]
)
```

```python
lemmatizer=WordNetLemmatizer()
df["tokens"]=df["tokens"].apply(
    lambda x:[lemmatizer.lemmatize(word)for word in x]
)
```

```python
df["tokens"]=df["tokens"].apply(
    lambda x:[word for word in x if len(word)>2]
)
```

```python
df["clean_sentence"]=df["tokens"].apply(lambda x: "".join(x))
```
